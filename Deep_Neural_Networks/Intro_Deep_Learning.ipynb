## Introduction to Deep Learning

### What is Deep Learning?

Deep Learning is a subset of Machine Learning (which is part of Artificial Intelligence) that uses Artificial Neural Networks to extract/learn complex patterns from large amounts of data.

### Introduction to the Artificial Neural Networks (ANN)

Artificial Neural Networks, ANN for short,are a computational model that was designed to mimic the human brain. Just like the brain is made up of neurons to process data and make decisions, ANNs consist of a network of artificial neurons that extract complex patterns from data and use that to make predictions.

**The Perceptron**

The basic unit of an ANN is the perceptron. Simply put, the perception receives a series input, calculates the weighted sum of the inputs, adds a bias term and then applies an activation function to this summation.



#### **Key Components of an ANN**

- **Input Layer**

Feeds data into the netwwork. The no. of neurons in this layer corresponds to the no. of features in your data set. There are no computations that occur here.

- **Hidden layer**

This is the brain of the network. A network can have an arbitrary number of hidden layers - the more the hidden layers, the 'deeper' the network and the more complex patterns the network can learn from the data.

- **Output Layer**

This is where the network produces its final prediction. the number of neurons depends of the type of problem you are solving for,regression- 1 neuron, classification- no. of labels. It is at this layer that choosing the correct activation function becomes very critical, so pay attention to the section on this notebook outlining activation functions.



#### Activation Functions
What is an activation function and what role do they actually play in the ANN. 

Activation functions play a critical role in neural networks by introducing non-linearity into the model, which allows the network to learn complex patterns beyond simple linear relationships. After a neuron computes a weighted sum of its inputs and adds a bias, the activation function determines how much of this signal is passed to the next layer. Without activation functions, even very deep neural networks would behave like a single linear model and would be unable to model real-world phenomena such as images, speech, or text. 

Activation functions act like gates that control when neurons activate, and influence information flow.As a result, choosing appropriate activation functions is essential for effective training of an ANN.


#### Types of Activation Functions

- Linear Activation
 This simply returns the input (x) as the output.
 ``` f(x)=x```
 its graphically represented as a straight line with a slope of 1. It is used in the output layer of a regression problem, never used in the hidden layer.



- Sigmoid Activation
Based on the sigmoid function, which takes any real number and squishes it between 0 and 1. These outputs (0 & 1), can be naturally be interpreted as probabilities and thus the sigmoid activation function is used in the output layer of  binary classification problems.



- Tanh (Hyperbolic tanget) Activation
The tanh function is similar to the sigmoid, only that this time it squiches any real value between -1 and 1. Generally, used in the hidden layer of neural networks and leads to faster convergence than sigmoid.



- Rectilinear Unit Activation (ReLU)
Simply what the ReLu does is, if the value is positive it keeps it, otherwise it makes it a zero. Best used in the hidden layers- if unsure of which activation to use in the hidden layer- this is usually the go-to choice.



- Leaky Rectilinear Unit Activation (Leaky ReLu)
This is a clever twist on the regular ReLU. While ReLU sets all negative inputs to zero, Leaky ReLU allows a small, non-zero, constant output for negative inputs



- Softmax Activation
Used in the output layer of a muti-class classification problem. Converts any real values into probability distribution which sums to 1.



### How do Neural Networks Actually Learn?

**Forward Propagation --> Loss function --> Back propagation --> Update parameters (Gradient Descent)**

Neural networks like many ML models learn by iteratively adjusting its parameters to reduce their cost function as much as possible (this is just the difference between what they predict and the correct answer)
In ANN the learning process can be divided into 2 clear phases, as discussed below...

##### Forward Propagation

What we have covered so far, in the sections above is actually the forward propagation phase of an ANN training. Foward propagation is the process of taking an input, passing it through the network, layer by layer and then making a prediction.

The same process - calculating the weighted sum of inputs, adding a bias term to this sum and then applying an activation function is repeated across all neurons in each layer. The output of the activation function is the input of the next layer of neurons.

#### Back Propagation

After the ANN has produced an output, we need a way to evaluate its performance. Next a loss function is computed as a feedback signal to how 'off' the model's prediction is, the loss function is simply

``` y predicted - y observed```

Backpropagation is the process of using the calculated Loss to determine precisely how every single weight and bias in the network should be adjusted to reduce that loss function in a future iteration. In essence, what back propagation seeks to do is 'assign' responsibility for the model's error to each parameter (weight and bias) and also determine the degree to which each parameter should change to reduce the loss function.

After we get the gradient of the loss function with respect to each parameter, we use gradient descent to efficiently update the values of the weights + biases and then the process starts again until convergence.
