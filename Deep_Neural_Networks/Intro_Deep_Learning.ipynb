## Introduction to Deep Learning

### What is Deep Learning?

Deep Learning is a subset of Machine Learning (which is part of Artificial Intelligence) that uses Artificial Neural Networks to extract/learn complex patterns from large amounts of data.

### Introduction to the Artificial Neural Networks (ANN)

Artificial Neural Networks, ANN for short,are a computational model that was designed to mimic the human brain. Just like the brain is made up of neurons to process data and make decisions, ANNs consist of a network of artificial neurons that extract complex patterns from data and use that to make predictions.

Let's go over the basic structure of an ANN below

#### Structure of ANNs

##### 1.The Perceptron

The basic unit of an ANN is the perceptron. Simply put, the perception receives a series input, calculates the weighted sum of the inputs, adds a bias term and then applies an activation function to this summation.

#### Activation Functions
What is an activation function and what role do they actually play in the ANN. 

Activation functions play a critical role in neural networks by introducing non-linearity into the model, which allows the network to learn complex patterns beyond simple linear relationships. After a neuron computes a weighted sum of its inputs and adds a bias, the activation function determines how much of this signal is passed to the next layer. Without activation functions, even very deep neural networks would behave like a single linear model and would be unable to model real-world phenomena such as images, speech, or text. 

Activation functions act like gates that control when neurons activate, and influence information flow.As a result, choosing appropriate activation functionssuch as ReLU for hidden layers and sigmoid or softmax for output layers depending on the taskis essential for effective neural network learning.


#### Types of Activation Functions

- Linear Activation
- Sigmoid Activation
- Tanh (Hyperbolic tanget) Activation
- Rectilinear Activation (ReLU)
