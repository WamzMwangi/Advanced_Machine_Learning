## Introduction to Computer Vision

Computer Vision is a field of Artificial Intelligence that enables machines to understand, analyze, interpret and extract data from digital images and videos. It's a very fast emerging field in tech that aims to create algorithms that 'see' the world in pretty much the same way the human eyes do - you know, where you are just not seeing, but you are seeing in context, and interpretting what you are seeing in a given context. It is very fascinating!




### Applications of Computer Visions

- **Object classification**: A system using object classification can categorize objects in an image based on predefined labels. For example, it can differentiate between people, animals, and vehicles. This helps with applications like traffic monitoring and inventory management.

- **Object detection and recognition**: The system can locate specific objects within an image or video and identify them. This is used in face recognition, crop disease detection and in diagnosing medical conditions from scans.

- **Object tracking**: The system can track the movement of objects by analyzing video frames over time. This is useful for autonomous vehicles, security surveillance, and sports performance analysis, I have also seen ground-breaking applications of this in wildlife monitoring

- **Optical character recognition (OCR)**: OCR converts text in images, scanned documents, and videos into digital text. It can process printed and handwritten text, though accuracy might depend on the quality of handwriting. OCR supports applications in document automation (like digitizing paper records), also, just thinking out loud here, imagine if we could have an app that nairobi water agents could use for meter reading.

- **Image and video segmentation**: Segmentation divides an image into distinct regions, which allows the system to recognize individual objects and their boundaries. This is important for self-driving cars, medical imaging, environmental monitoring.

- **3D object recognition and depth perception**: Some computer vision systems analyze depth and spatial relationships to recognize objects in three dimensions. This is essential for robotics, augmented reality and virtual reality experiences, and industrial automation.

- **Image generation and enhancement**: Computer vision can generate, restore, and enhance images. This can improve photo resolution, remove noise, and even create synthetic images for training AI models.

## How do Computers 'See'?

**How is it that a machine is able to 'see' and image? What does a computer see when it 'looks' at an image or video input?**

Every image can actually be defined as its pixel representation.Pixels are the basic building blocks of a digital image. A pixel is what we call the color or light values that occupy a specific place in an image. Think of an image as a big grid, with each square in the grid containing one color or pixel. An image with a resolution of 1024768 is a grid with 1,024 columns and 768 rows, which therefore contains 1,024  768 = 786,432 pixels.

**Okay so, now we know what an image is in the eyes of a computer, its just a grid of numbers, how is it that a machine can be able to tell a car from a human face?**

Simply put, for an computer vision algorithm to be able to tell apart a car from face, it has to extract and learn the features that make a car, a car and the features that make a face, a face! So for example, it has to learn that a car has wheels, windscreens,bonnet etc... and then learn that a human face has eyes, ears, nose etc...From this, we see that a really big task in computer vision is feature extraction!

**How Deep Neural Networks Extract features in Computer Vision**

Deep neural networks allow us to implicitely extract and learn from images. It does this in a hierarchiacal manner. The network first learns low level features, then mid-level and then high level features. For example, the first hidden layer will learn the edges and dark spots, then the next layer will learn how these edges come together to form the edges of the nose, mouth eyes and until finally we get to deeper layers that now learn the entire facial structure.



### Computer Vision with Fully Connected Neural Networks

The ANN architecture we applied so far is actually called a ***'fully connected neural network'*** - this just means every neuron in the network is connected to another neuron.

Fully connected neural networks can be used in computer vision tasks but with a slight modification on how the image input is fed into the input layer. Remember, an image we have said is represented as a grid of pixel values- this means an image is 2-dimensional. How can we feed 2D data into the 1D input layer?

We do something called flattenning -this is where the 2D image vector is converted into a long 1D vector before it is fed into the iput layer.

### Computer Vision with Convolutional Neural Networks

When you see an image, the visual processing centre of the brain does not necessarily interpret the ENTIRE image at once, it might feel like we do, but not quite. What happens is each neuron in the visual processing centre, processing visual data in its receptive field and then it connects to other neurons processing their respective receptive fields, until eventually your vision centre covers the entire visual field.

Just like the human brain, a convolutional neural network extracts data from an image in a step-wise process. Instead of looking at the entire image like in a FCNN, a CNN looks at small regions at a time and extracts features from an image layer by layer. The way a CNN does this is through an operation called a **convolution**. This way a CNN preserves the spatial information in an image and reduces the number of parameters in your network.

### Components of a Convolutional Neural Network

#### 1. Convolutional Layers

In the convolution layer, we use small grids (called filters or kernels) that move over the image. Each small grid is like a mini magnifying glass that looks for specific patterns in the photo, like lines, curves, or shapes. As it moves across the photo, it creates a new grid that highlights where it found these patterns.

For example, one filter might be good at finding straight lines, another might find curves, and so on. By using several different filters, the CNN can get a good idea of all the different patterns that make up the image.



#### 2. Activation Function
A ReLU activation function is applied after each convolution operation. This function helps the network learn non-linear relationships between the features in the image, hence making the network more robust for identifying different patterns.



#### 3. Pooling Layer
The goal of the pooling layer is to pull the most significant features from the convoluted matrix. This is done by applying some aggregation operations, which reduce the dimension of the feature map (convoluted matrix), hence reducing the memory used while training the network. Pooling is also relevant for mitigating overfitting.



#### 4. Fully Connected Neural Network Layer
The last layer of the pooling is what is flattenned and fed as input into the fully connected neural network
